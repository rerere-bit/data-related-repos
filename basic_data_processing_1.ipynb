Simple data processing with a case study: Data analysis of student grades in various majors, calculating final grades (assignment grade * 0.6 + exam grade * 0.4) and student status.

```python
#Importing Library

import pandas as pd #Library for working with dataset
import numpy as np
```

```python
#Declarating Data

raw_data = {
    'nama': ['Andi', 'Budi', 'Citra', 'Dedi', 'Eka', 'Fani'],
    'jurusan': ['Informatika', 'Sistem Informasi', 'Informatika', 'Informatika', 'Sistem Informasi', 'DKV'],
    'nilai_tugas': [80, 90, 75, np.nan, 85, 60],
    'nilai_ujian': ['85', '70', '90', '60', '88', '50'],
    'kehadiran': [100, 90, 80, 50, 95, 40]
}
```

```python
df = pd.DataFrame(raw_data) #Mengubah dictionary menjadi tabel (data frame)

print('--- Tampilan Awal Data ---')
display(df)
```

```python
#Inspecting Data

df.info()
```

```python
#Cleaning

df['nilai_tugas'] = df['nilai_tugas'].fillna(0) #Filling empty value (Nan) with 0
df['nilai_ujian'] = df['nilai_ujian'].astype(int) #Converting data type into int
```

```python
#Simple Math Manipulation

df['nilai_akhir'] = (df['nilai_tugas'] * 0.6) + (df['nilai_ujian'] * 0.4)
df['status'] = np.where(df['nilai_akhir'] > 60, 'LULUS', 'REMEDIAL')
```

```python
#Grouping

df.groupby('jurusan')['nilai_akhir'].mean()
```

```python
#Displaying Result

display(df)
```

Scenario: We want to see all informatics student that passed.

```python
#Filtering

df_informatika = df[df['jurusan'] == 'Informatika']
df_informatika_lulus = df_informatika[df_informatika['status'] == 'LULUS']
print(df_informatika_lulus)
```
